```python
import os
import pandas as pd

# Dossier de base où se trouvent les fichiers par date
BASE_PATH = 'local_data/globalshop-raw'

# Stocker tous les DataFrames ici
all_dfs = []

# Parcourir chaque dossier de date
for date_folder in os.listdir(BASE_PATH):
    date_path = os.path.join(BASE_PATH, date_folder)
    if os.path.isdir(date_path):
        for filename in os.listdir(date_path):
            if filename.endswith('_orders.csv'):
                file_path = os.path.join(date_path, filename)
                df = pd.read_csv(file_path)

                # Ajouter la colonne de canal à partir du nom du fichier
                if filename.startswith('WEB'):
                    df['channel'] = 'WEB'
                elif filename.startswith('PART'):
                    df['channel'] = 'PART'
                elif filename.startswith('MOB'):
                    df['channel'] = 'MOB'

                all_dfs.append(df)

# Regrouper tous les fichiers dans un seul DataFrame
df = pd.concat(all_dfs, ignore_index=True)

# Calculer la colonne revenue
df['revenue'] = df['quantity'] * df['unit_price']

# Vérification
print(f"{len(df)} lignes chargées")
print(df.head())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.to_period('M')

df['revenue'] = df['quantity'] * df['unit_price']
monthly_country_revenue = df[df['status'] == 'PAID'].groupby(['month', 'country'])['revenue'].sum().unstack()

monthly_country_revenue.plot(kind='bar', figsize=(14, 6), title="Chiffre d'affaires mensuel par pays")
```

```python
# En valeur
top_value = df[df['status'] == 'PAID'].groupby('product_id')['revenue'].sum().nlargest(10)

# En quantité
top_quantity = df[df['status'] == 'PAID'].groupby('product_id')['quantity'].sum().nlargest(10)

# Visualisation
top_value.plot(kind='bar', title="Top 10 produits (valeur)")
top_quantity.plot(kind='bar', title="Top 10 produits (quantité)")
```

```python
cancel_rate = df.groupby('country')['status'].value_counts(normalize=True).unstack().fillna(0)
cancel_rate['CANCELLED'].plot(kind='bar', title="Taux d'annulation par pays", figsize=(10, 5))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supposons que df est déjà construit avec toutes les données fusionnées
df['order_date'] = pd.to_datetime(df['order_date'])
df['revenue'] = df['quantity'] * df['unit_price']
df['site'] = df['order_id'].str.extract(r'^(WEB|PART|MOB)')

# Groupe par jour et site
daily_revenue = df[df['status'] == 'PAID'].groupby(['order_date', 'site'])['revenue'].sum().unstack()

# Visualisation
daily_revenue.plot(marker='o', figsize=(12, 6), title="Évolution quotidienne du chiffre d'affaires par site")
plt.ylabel("Chiffre d'affaires")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()
```
